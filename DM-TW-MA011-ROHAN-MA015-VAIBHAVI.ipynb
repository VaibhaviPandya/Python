#### Apriori algorithm
<li> Use the apriori class from the apyori library.
<li> The apriori class requires some parameter values to work. 
<li> First parameter is the list of list that you want to extract rules from.
<li> Second parameter min_support: to select the items with support values greater than min_support.
<li> Third parameter min_confidence: filters those rules that have confidence greater than the confidence threshold min_confidence.
<li> Fourth paramter min_lift: specify the minimum lift value for the short listed rules.
<li> Fifth  parameter min_length: specifies the minimum number of items that are required in the generated rules.

```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv('netflix_titles.csv')
```

remove header name

```python
df = pd.read_csv('netflix_titles.csv', header=None)
df
```

Drop nan value from dataset

```python
df = df.dropna()

print (df)
```

```python
df.shape
```

#### Data presentation

Transform dataset to be in the form of a list of lists-<br>
<li> Currently the available data is in the form of a pandas dataframe. 
<li> The whole dataset should be a big list and each transaction in the dataset is an inner list within the outer big list. 

```python
records = []
for i in range(0, df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.045, min_confidence=0.2, min_lift=2, min_length=1)


# min_support is 0.0045:
# Find rules for only those items that are purchased at least 5 times a day i.e. 7 days X 5 times a day = 35 times a week
# Given dataset is for one week.
# Hence, min. support = 35 / 7500 = 0.0045

# min_confidence = 20% i.e. 0.2
# min_lift = 3
# min_length = 2 
```

```python
association_rules
```

```python
association_results = list(association_rules)
print(association_results)
```
